```python
# ========================================================
# NOTEBOOK 01 - AN√ÅLISIS EXPLORATORIO DE DATOS (EDA)
# ========================================================
# Este cuaderno realiza una exploraci√≥n general del dataset
# "Online Retail", detecta valores nulos y duplicados,
# visualiza distribuciones y deja una versi√≥n limpia del dataset
# para los modelos de clustering.
```

```python
# ===============================================
# BLOQUE 1 - Importaci√≥n de librer√≠as necesarias
# ===============================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configuraci√≥n visual de gr√°ficos
sns.set(style="whitegrid")
```

```python
# ===============================================
# BLOQUE 2 - Carga del dataset original
# ===============================================

# Cargamos el archivo previamente convertido a CSV
df = pd.read_csv('../data/online_retail.csv')

# Confirmamos forma del dataset
print(f"‚úÖ Dataset cargado correctamente. Dimensiones: {df.shape}")
```

```python
# ===============================================
# BLOQUE 3 - Estructura general del dataset
# ===============================================

print("üìå Columnas disponibles:")
print(df.columns.tolist())

print("\nüîç Tipos de datos:")
print(df.dtypes)
```

```python
# ===============================================
# BLOQUE 4 - Revisi√≥n de valores nulos y duplicados
# ===============================================

print("üö´ Valores nulos por columna:")
print(df.isnull().sum())

print(f"\nüìõ Registros duplicados: {df.duplicated().sum()}")
```

```python
# ===============================================
# BLOQUE 5 - Visualizaci√≥n de valores faltantes
# ===============================================

plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Mapa de calor de valores nulos")
plt.tight_layout()
plt.show()
```

```python
# ===============================================
# BLOQUE 6 - Limpieza de datos
# ===============================================

# Eliminamos registros sin CustomerID y hacemos copia segura
df_clean = df.dropna(subset=['CustomerID']).copy()

# Convertimos el ID de cliente a entero
df_clean['CustomerID'] = df_clean['CustomerID'].astype(int)

# Eliminamos columnas no √∫tiles para clustering
df_clean.drop(columns=['InvoiceNo', 'StockCode', 'Description', 'InvoiceDate'], inplace=True)

# Confirmamos estructura limpia
print(f"‚úÖ Dataset limpio. Filas: {df_clean.shape[0]} | Columnas: {df_clean.columns.tolist()}")
```

```python
# ===============================================
# BLOQUE 7 - Distribuci√≥n de cantidad de productos
# ===============================================

plt.figure(figsize=(8, 4))
sns.histplot(df_clean['Quantity'], bins=50, kde=True)
plt.title("Distribuci√≥n de Cantidad de Productos Comprados")
plt.tight_layout()
plt.show()
```

```python
# ===============================================
# BLOQUE 8 - Distribuci√≥n del precio unitario
# ===============================================

plt.figure(figsize=(8, 4))
sns.histplot(df_clean['UnitPrice'], bins=50, kde=True)
plt.title("Distribuci√≥n del Precio Unitario")
plt.tight_layout()
plt.show()
```

```python
# ===============================================
# BLOQUE 9 - Pa√≠ses con m√°s transacciones
# ===============================================

top_paises = df_clean['Country'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_paises.values, y=top_paises.index)
plt.title("Top 10 pa√≠ses con m√°s transacciones")
plt.xlabel("N√∫mero de transacciones")
plt.ylabel("Pa√≠s")
plt.tight_layout()
plt.show()
```

```python
# ===============================================
# BLOQUE 10 - Guardar el dataset limpio
# ===============================================

# Guardamos una copia del dataset limpio para los modelos
df_clean.to_csv('../data/online_retail_clean.csv', index=False)
print("üìÅ Dataset limpio guardado como: online_retail_clean.csv")
```
