```python
# ========================================================
# NOTEBOOK 03 - VISUALIZACIÓN DE CLUSTERS Y ANÁLISIS FINAL
# ========================================================
# Este cuaderno aplica reducción de dimensionalidad con PCA
# para visualizar los clusters obtenidos con K-means y DBSCAN.
# También genera una tabla resumen de perfiles por cluster.
```

```python
# ===============================================
# BLOQUE 1 - Importación de librerías
# ===============================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# ===============================================
# BLOQUE 2 - Carga del dataset normalizado con clusters
# ===============================================

df_norm = pd.read_csv('../data/online_retail_clean.csv')  # si guardaste aquí también los clusters, reemplaza por correcto
# Si df_norm con clusters ya está creado en memoria, puedes usarlo directamente
```

```python
# ===============================================
# BLOQUE 3 - Visualización de clusters K-means con PCA
# ===============================================
# Este bloque aplica PCA sobre las variables normalizadas
# y muestra visualmente cómo se agrupan los clientes
# según los clusters generados con K-means.

# Cargamos el dataset con normalización y clusters ya procesados
df_norm = pd.read_csv('../data/online_retail_norm_clusters.csv')

# Seleccionamos las columnas necesarias para PCA
X = df_norm[['TotalProductos_norm', 'PrecioPromedio_norm', 'PaisesComprados_norm']]

# Aplicamos PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
componentes = pca.fit_transform(X)

# Creamos DataFrame con componentes
df_pca = pd.DataFrame(componentes, columns=['PCA1', 'PCA2'])
df_pca['cluster_kmeans'] = df_norm['cluster_kmeans'].astype(str)

# Visualización y guardado
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='cluster_kmeans', palette='Set2', s=60)
plt.title("Visualización 2D de Clusters - K-means (PCA)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.savefig('../results/graficos/pca_kmeans_clusters.png')  # Guardar antes del show
plt.show()

```

```python
# ===============================================
# BLOQUE 4 - Visualización de clusters DBSCAN con PCA
# ===============================================
# Este bloque utiliza las componentes principales generadas por PCA
# para visualizar gráficamente los agrupamientos encontrados por DBSCAN.
# Los puntos con etiqueta "-1" representan ruido o clientes no agrupados.

# Verificamos si ya existe df_pca con resultados de PCA
if 'df_pca' not in locals():
    raise ValueError("❌ Error: df_pca no está definido. Ejecuta el BLOQUE 3 primero.")

# Confirmamos que cluster_dbscan exista
if 'cluster_dbscan' not in df_norm.columns:
    raise ValueError("❌ La columna 'cluster_dbscan' no está disponible. Verifica el cuaderno 02.")

# Agregamos la etiqueta de cluster DBSCAN al dataframe visual
df_pca['cluster_dbscan'] = df_norm['cluster_dbscan'].astype(str)

# Visualización
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='cluster_dbscan', palette='Set1', s=60)
plt.title("Visualización 2D de Clusters - DBSCAN (PCA)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```

```python
# Guarda gráfico K-means
plt.savefig('../results/graficos/pca_kmeans_clusters.png')

# Guarda gráfico DBSCAN
plt.savefig('../results/graficos/pca_dbscan_clusters.png')
```

```python
# ===============================================
# BLOQUE 5 - Tabla resumen de perfiles por cluster (K-means)
# ===============================================
# Este bloque reconstruye las métricas por cliente
# y las combina con los clusters asignados previamente
# para generar una tabla de perfiles promedio por grupo.

# Cargamos el dataset limpio original
df_clean = pd.read_csv('../data/online_retail_clean.csv')

# Reagrupar por cliente
df_grouped = df_clean.groupby('CustomerID').agg({
    'Quantity': 'sum',
    'UnitPrice': 'mean',
    'Country': 'nunique'
}).reset_index()

df_grouped.rename(columns={
    'Quantity': 'TotalProductos',
    'UnitPrice': 'PrecioPromedio',
    'Country': 'PaisesComprados'
}, inplace=True)

# Cargamos el dataset con clusters
df_clusters = pd.read_csv('../data/online_retail_norm_clusters.csv')

# Combinamos ambos DataFrames por CustomerID
df_perfiles = df_grouped.merge(df_clusters[['CustomerID', 'cluster_kmeans']], on='CustomerID')

# Agrupamos por cluster y resumimos
tabla_perfiles = df_perfiles.groupby('cluster_kmeans').agg({
    'TotalProductos': 'mean',
    'PrecioPromedio': 'mean',
    'PaisesComprados': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'NumClientes'}).round(2)

# Mostramos tabla resumen
print("📋 Tabla resumen de características promedio por cluster (K-means):")
display(tabla_perfiles)

# Exportar tabla resumen
tabla_perfiles.to_csv('../results/tablas/tabla_perfiles_kmeans.csv')
```
